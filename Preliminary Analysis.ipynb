```python
import pandas as pd

base = pd.read_csv('dataset\Base.csv')
variant1 = pd.read_csv('dataset\Variant I.csv')
variant2 = pd.read_csv('dataset\Variant II.csv')
variant3 = pd.read_csv('dataset\Variant III.csv')
variant4 = pd.read_csv('dataset\Variant IV.csv')
variant5 = pd.read_csv('dataset\Variant V.csv')
```

```python
import matplotlib.pyplot as plt
fig1,axes1=plt.subplots(1,3,figsize=(3,3),dpi=144)
fig2,axes2=plt.subplots(1,3,figsize=(3,3),dpi=144)



for ax, df, title in zip(axes1,(base, variant1, variant2), ('base', 'variant1', 'variant2')):
    count = df.fraud_bool.value_counts().to_frame().sort_index()
    ax.pie(count.fraud_bool, labels=count.index)
    ax.set_title(title)

for ax, df, title in zip(axes2,(variant3, variant4, variant5), ('variant3', 'variant4', 'variant5')):
    count = df.fraud_bool.value_counts().to_frame().sort_index()
    ax.pie(count.fraud_bool, labels=count.index)
    ax.set_title(title)
```

```python
base_count = base.fraud_bool.value_counts().to_frame().transpose()
variant1_count = variant1.fraud_bool.value_counts().to_frame().transpose()
variant2_count = variant2.fraud_bool.value_counts().to_frame().transpose()
variant3_count = variant3.fraud_bool.value_counts().to_frame().transpose()
variant4_count = variant4.fraud_bool.value_counts().to_frame().transpose()
variant5_count = variant5.fraud_bool.value_counts().to_frame().transpose()
```

```python
count = pd.concat([base_count,variant1_count,variant2_count,variant3_count,variant4_count,variant5_count])
count.index=['base', 'variant1', 'variant2','variant3', 'variant4', 'variant5']
```
